## 1. Import Libraries & Driver

````python
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
import time
````

## 2. EXTRACT DATA

````python
# Mulai jalankan browser
driver = webdriver.Edge()
url = 'https://matahari.com/collections/desember-new-arrival?filter.v.m.custom.discount_group=10%25+ke+atas'

# Penarikan konten dengan infinite scroll
driver.get(url)
for i in range(5):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2) 

# Penarikan source HTML setelah konten ditampilkan
html = driver.page_source 
soup = BeautifulSoup(html, "html.parser")

# Pengambilan elemen berdasarkan tags, attribute, dan attribute_value
el_product = soup.find_all('div', {'class': 'grid-product__title'})
el_price = soup.find_all('div', {'class': 'grid-product__price'})
el_brand = soup.find_all('div', {'class': 'grid-product__vendor'})
el_diskon = soup.find_all('span', {'class': 'grid-product__price--savings'})

# List untuk menampung data mentah
product_name = []
price = []
brand = []
diskon = []

# Iterasi dari tiap elemen
for data in el_product:
    product_name.append(data.get_text())

for data in el_price:
    price_text = data.find('span').get_text().strip().strip()
    price.append(price_text)

for data in el_brand:
    brand_name = data.get_text().strip()
    brand.append(brand_name)

for data in el_diskon:
   diskon.append(data.get_text().strip().replace('-', ''))

# Menghentikan driver
driver.quit

# Konversi hasil ekstrak ke Df Pandas
df = pd.DataFrame({
    'brand' : brand,
    'product_name' : product_name,
    'diskon' : diskon,
    'price' : price})
````

````python
# Melihat isi konten df
df
````

````python
# Menyimpan data hasil ekstraksi ke csv
df.to_csv('dataset_milestone.csv')
````

## 3. TRANSFORM

````python
# Memuat data dari file CSV ke dalam DataFrame pandas
df = pd.read_csv('dataset_milestone.csv')
````

````python
# Menghapus kolom yang tidak diperlukan
df = df.drop(columns='Unnamed: 0')
````

### Cleaning kolom "diskon"

````python
# Mengambil angkanya saja dengan menghilangkan '%' dan
df['diskon'] = df['diskon'].str.replace('%', '')

# Mengganti tipe data kolom menjadi int
df['diskon'] = df['diskon'].astype(int)
````

### Cleaning kolom "price"

````python
# Mengambil angkanya saja dengan menghilangkan 'Rp' dan '.'
df['price'] = df['price'].str.replace('Rp', '').str.replace('.', '')

# Mengganti tipe data kolom menjadi int
df['price'] = df['price'].astype(int)
````
